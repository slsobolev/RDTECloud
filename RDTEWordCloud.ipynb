```python
df = pd.read_csv('RDTE_Text.csv')
df.head()
```

```python
budgetactivity = df.loc[df['Budget Activity Number'] == 6]   
plantext = budgetactivity['Accomplishment Planned Program Current Year Text'].dropna()
```

```python
text = ''
for t in plantext:
    text = text + ' ' + str(t) 
text = text.lower()
```

```python
listext = text.split()
```

```python
dodstop = ['dod','demonstrate','transition','tools','begin','prototype','upgrade','evaluate','plan','demonstration','based','tool','activities','planning','component','threat','environment','assessment','including','mission','operation','architecture','management','advanced','research','evaluation','efforts','performance','engineering','improve','technologies','data','test','platform','new','model','requirements','include','software','technologies','integration','effort','testing','nan','continue','develop','support','capability','capabilities','provide','program','analysis','conduct','system','complete','development','requirement','systems','will','initiate','design','technology']

stopw = set(list(wordcloud.STOPWORDS) + listext[:200] + dodstop)
print len(wordcloud.STOPWORDS)
print len(stopw)
```

```python
wc = wordcloud.WordCloud(stopwords=stopw,max_words=300,width=1200,height=800, background_color='white', margin=10)
                         
timestamp = str(time.time())
wc.generate(text)
wc.to_file('rdtecloud ' + timestamp + '.png')
```
